```python
#!/usr/bin/env python
# coding: utf-8

# In[8]:


import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter
from scipy.signal import freqz

#Define the coefficients
b = [0,-23,0,105,0,-263,0,526,0,-949,0,1672,0,-3216,0,10342,16384,10342,0,-3216,0,1672,0,-949,0,526,0,-263,0,105,0,-23]

#Normalize the coefficients by dividing by 32768
b = np.array(b)/32768.0

print("Implemented coefficients:",b)

#Create time variable 
t = np.linspace(0,1,1000) #1 KHz
# print(t*1000)

#Generate input signals
signal = np.sin(2 * np.pi *t * 0.5*(1000.0)) + np.sin(2 * np.pi *t * (1000.0))

#Define a filter function 
def shannon_whitaker_filter(signal,b):

#This function will apply the low-pass filter to the input signal

    filtered_signal = lfilter(b, [1.0],signal)

    return filtered_signal

filtered_signal = shannon_whitaker_filter(signal,b)

plt.figure(figsize = (10,6))
plt.plot(t,signal,label = 'Input signal')
plt.plot(t,filtered_signal,label = 'Filtered signal')
plt.legend()
plt.title('Shannon-Whitaker Low-Pass filter')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.show()

#Compute the frequency response of the filter
w,h = freqz(b,a=1)

# Plot the amplitude response
plt.figure(figsize=(10, 6))
plt.plot( w, 20 * np.log10(abs(h)), 'b')
plt.title('Frequency Response of the Shannon-Whitaker Low-Pass Filter')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (dB)')
plt.grid()
plt.show()


# In[99]:


import numpy as np
import matplotlib.pyplot as plt

# Create time variable 
t = np.linspace(0, 1, 1000)

signal1 = np.sin(2 * np.pi * 2000 * t)

plt.plot(t, signal1)  # Specify the x-axis values
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('1 Hz Sine Wave')

plt.show()


# In[96]:


import numpy as np
import matplotlib.pyplot as plt

# Create time variable 
t = np.linspace(0, 1, 500)

signal2 = np.sin(2 * np.pi * 5 * t)

plt.plot(t, signal2)  # Specify the x-axis values
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('5 Hz Sine Wave')

plt.show()



# In[100]:


import numpy as np
import matplotlib.pyplot as plt

# Create time variable 
t = np.linspace(0, 1, 500)

signal3 = np.sin(2 * np.pi * 5 * t)+ np.sin(2 * np.pi * 1 * t)

plt.plot(t, signal3)  # Specify the x-axis values
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Input signal')

plt.show()


# In[ ]:




```
